```python
import os
import pandas as pd

# Define the folder path
folder_path = r'E:\Python\Test\AD Case Study\Datasets'

# List to store DataFrames
dataframes = []

# Loop through files in the folder
for file_name in os.listdir(folder_path):
    # Check if the file starts with 'MONTH_' and ends with '.csv'
    if file_name.startswith('MONTH_') and file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        dataframes.append(df)

# Combine all DataFrames into one
combined_df = pd.concat(dataframes, ignore_index=True)

print(combined_df['MONTH'].unique())
print(len(combined_df))

combined_df.head()
```

```python
df_account_performance = pd.read_csv(r'E:\Python\Test\AD Case Study\Datasets\ACCOUNT_AGE_PERFORMANCE.csv')
df_merged = pd.merge(combined_df,df_account_performance,on='ACCOUNT_ID',how='left')

# Function to assign bands
def assign_band(seats):
    if seats == 1:
        return '1 Seat'
    elif 2 <= seats <= 5:
        return '2-5 Seats'
    elif 6 <= seats <= 10:
        return '5-10 Seats'
    elif 11 <= seats <= 50:
        return '10-50 Seats'
    elif 51 <= seats <= 100:
        return '50-100 Seats'
    elif 101 <= seats <= 200:
        return '100-200 Seats'
    elif 201 <= seats <= 500:
        return '200-500 Seats'
    else:
        return '500+ Seats'

# Apply the function to create a new column
df_merged['Customer Size Band'] = df_merged['SEATS_PURCHASED'].apply(assign_band)
print(len(df_merged))

df_merged.to_csv(r'E:\Python\Test\AD Case Study\Datasets\Final.csv', index=False)
df_merged.head()
```
